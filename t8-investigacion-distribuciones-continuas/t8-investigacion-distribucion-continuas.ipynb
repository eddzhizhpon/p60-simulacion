## A. Introducción

Las distribuciones continuas de probabilidades son muy útiles cuando se emplean la representación de modelos estocásticos para procesos que requieran de una variable aleatoria continua. Una variable aleatoria continua es una variable aleatoria con un conjunto de valores posibles (conocido como el rango) que es infinito y no se puede contar.

A continuación, se realizará una descripción de técnicas para generar valores de variables aleatorias a partir de distribuciones continuas: Pareto, T de Student, Chi cuadrado, Beta, Gamma y Log-normal. Finalmente, se darán a conocer las funciones en Python necesarias para la generación de valores aleatorios y ejemplo orientado a la realidad.

## B. Descripción

### 1. Pareto

Esta distribución fue introducida por su inventor Vilfredo Pareto. El objetivo es explicar la distribución de los salarios en la sociedad. A menudo, se describe como la base de la regla 80%/20%. Por ejemplo, el 80% de las quejas de clientes con respecto al funcionamiento del vehículo surgen del 20% de los componentes.

La distribución de Pareto está dada por la función:

$$
p(x;\alpha, k) = \frac{\alpha k^{\alpha}}{x^{\alpha + 1}}
$$

Consideraciones:

- $x \geq k$ y $\alpha > 0$ son Números Reales

### 2. T de Student

Esta distribución es utilizada para probar si la diferencia entre las medias de dos muestras de observaciones es estadísticamente significativa. 

Por ejemplo, las alturas de una muestra de jugadores de baloncesto podrían compararse con las alturas de una muestra aleatoria de jugadores de fútbol. Esta distribución nos podría ayudar a determinar si un grupo es significativamente más alto que el otro.

La distribución t de Student está dada por la función:

$$
p(t;n) = \frac{ \Gamma \left( \frac{ n + 1 } { 2 } \right) } { \sqrt{n\pi} \Gamma \left( \frac{ n } { 2 } \right) } \left( 1 + \frac{ t^{2} } { 2 } \right)^{- \frac{ n + 1 } { 2 }}
$$

Consideraciones: 

- $t$ es un número real
- $n$ es un número entero positivo

### 3. Chi cuadrado

La aplicación de Chi cuadrado es que con la distribución se puede saber cuánto se ajustan los valores predichos para un conjunto de datos de un modelo teórico.

La distribución de Chi Cuadrado está dada por la función:

$$
p(x;n) = \frac{ \left( \frac{ x } { 2 } \right)^{ \frac{ n } { 2 } -1 } e \frac{ -x } { 2 } } { 2\Gamma \left(  \frac{ n } { 2 } \right) }
$$

Donde:

- La variable $x \geq 0$ y el parámetro $n$ es un número entero positivo
- El grado de libertad es un número entero positivo


### 4. Beta

Esta distribución puede representar variables aleatorias que tienen un rango finito de $a$ hasta $b$. 

Por ejemplo, la Distribución Beta se utiliza también con frecuencia como una probabilidad a priori para proporciones binomiales en el análisis bayesiano.

La distribución Beta está dada por la fórmula:

$$
p(x;p,q) = \frac{ 1 } { B(p,q) } x^{p-1}(1-x)^{q-1}
$$

Consideraciones:

- Los parámetros $p$ y $q$ son números reales positivos
- La variable $x$ satisface la condición $0 \leq x \leq 1$
- $B(p,q)$ es la función beta

### 5. Gamma

La Distribución Gamma comienza en el origen de coordenadas y tiene una forma bastante flexible. Otras distribuciones son casos especiales de ella.

La Distribución Gamma está dada por la fórmula:

$$
p(x;a,b) = \frac{ a(ax)^{ b-1 } e^{-ax} } { \Gamma(b) }
$$

Consideraciones:

- Los parámetros $a$ y $b$ son números reales positivos
- La variable $x$ es un número real positivo
- $\Gamma(b)$ es la función gamma

### 6. Log-normal

La Distribución Log-normal es aplicable a variables aleatorias que están limitadas por cero, pero tienen pocos valores grandes. Es una distribución con asimetría positiva. 

Algunos de los ejemplos en que la solemos encontrar son:
- El peso de los adultos.
- La concentración de los minerales en depósitos.
- Duración de licencia por enfermedad.
- Distribución de riqueza
- Tiempos muertos de maquinarias.

La Distribución Log-normal está dada por la fórmula:

$$
p(x;\mu;\sigma) = \frac{ 1 } { x\sigma\sqrt{2\pi} } e^{ \frac{ -1 } { 2 } \left( \frac{ ln\ x - \mu } { \sigma } \right)^{2} }
$$

Consideraciones:

- La variable $x > 0$
- Los parámetros $\mu$ y $\sigma$ son número reales

## C. Ejemplo del uso

### 1. Pareto

```python
# Graficando Pareto
k = 2.3 # parametro de forma.
pareto = stats.pareto(k)
x = np.linspace(pareto.ppf(0.01),
                pareto.ppf(0.99), 100)
pareto_pdf = pareto.pdf(x) # Función de Probabilidad
plt.plot(x, pareto_pdf)
plt.title('Distribución de Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# histograma
aleatorios = pareto.rvs(1000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```

**Ejemplo de aplicación**

Permite detectar los problemas que tienen más relevancia mediante la aplicación del principio de Pareto (pocos vitales, muchos triviales). Lo que se traduce a que hay muchos problemas sin importancia frente a unos graves. Por lo general, el 80% de los resultados totales se originan en el 20% de los elementos.

Asimismo, unos ejemplos prácticos serían:

- La minoría de clientes que representen la mayoría de las ventas.
- Una minoría de productos, procesos, o características de la calidad causantes del grueso de desperdicio o de los costos de reelaboración.
- La minoría de rechazos representa la mayoría de quejas de la clientela.
- Una minoría de vendedores está vinculada a la mayoría de partes rechazadas.
- La minoría de problemas causantes del grueso del retraso de un proceso.

### 2. T de Student

```python
df = 50
t = stats.t(df)
x = np.linspace(t.ppf(0.01),
                t.ppf(0.99), 100)
t_fp = t.pdf(x) # Función de Probabilidad
plt.plot(x, t_fp)
plt.title('Distribución t de Student')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# histograma
aleatorios = t.rvs(1000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma t de Student')
plt.show()
```

**Ejemplo de aplicación**

La distribución t se utiliza cuando:
- Queremos estimar la media de una población normalmente distribuida a partir de una muestra pequeña. 
- Tamaño de la muestra es inferior a 30 elementos, es decir, n < 30.

Por ejemplo, un estudio revelaba que la velocidad de lectura de niños entre 6 y 8 años de edad con Síndrome de Down que se habían atendido en el Centro de Apoyo para el Aprendizaje era de 74 palabras por minuto en lecturas seleccionadas. Se hizo un muestreo al azar con 24 niños de tales características para corroborar si se encontraban cerca de este patrón. Se obtuvo una media de 78 76 palabras por minuto con desviación estándar de 3 palabras por minuto. ¿Es el promedio de la muestra diferente al del patrón de 74 ppm?

Este problema puede ser resuelto utilizando la distribución t de Student, teniendo un valor de confianza, por ejemplo, 95%.

### 3. Chi cuadrado


```python
# Graficando Chi cuadrado

# parametro de forma.
df = 34
chi2 = stats.chi2(df)
x = np.linspace(chi2.ppf(0.01),
                chi2.ppf(0.99), 100)

# Función de Probabilidad
chi2_fp = chi2.pdf(x)
plt.plot(x, chi2_fp)
plt.title('Distribución Chi cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# Generación de valores aleatorios
aleatorios = chi2.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Chi cuadrado')
plt.show()
```

**Ejemplo de aplicación**

La prueba chi cuadrado se aplica a variables que describen cualidades, como por ejemplo sexo, estado civil, grupo sanguíneo, color de ojos y preferencias de diversos tipos.

La prueba está pensada cuando se desea:

- Comprobar si una distribución es apropiada para describir una variable, lo cual se denomina bondad del ajuste. Mediante la prueba de chi cuadrado se puede saber si hay diferencias significativas entre la distribución teórica seleccionada y la distribución de frecuencias observada.

Dado que se aplica a variables cualitativas o categóricas, la prueba chi cuadrado se utiliza ampliamente en ciencias sociales, administración y medicina.

Por ejemplo, supóngase que los investigadores desean conocer si la preferencia por el café negro está relacionada con el género de la persona, y especificar la respuesta con un nivel de significancia de $\alpha = 0.05$.

### 4. Beta

```python
# Graficando Beta

# Parametros de forma
a, b = 2.3, 0.6
beta = stats.beta(a, b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)

# Función de Probabilidad
beta_fp = beta.pdf(x)
plt.plot(x, beta_fp)
plt.title('Distribución Beta')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# Generación de valores aleatorios
aleatorios = beta.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Beta')
plt.show()
```

**Ejemplo de aplicación**

La distribución beta se usa generalmente para estudiar las variaciones, a través de varias muestras, de un porcentaje que representa algún fenómeno, por ejemplo, el tiempo diario que la gente dedica a mirar televisión.

Por ejemplo, Tanya ingresa a una rifa en la feria local y se pregunta cuáles son sus posibilidades de ganar. Si su probabilidad de ganar puede modelarse mediante una distribución beta con α = 5 y β = 2, ¿cuál es la probabilidad de que tenga como máximo un 10% de posibilidades de ganar?

### 5. Gamma

```python
# Graficando Gamma

# Parametro de forma
a = 2.6
gamma = stats.gamma(a)
x = np.linspace(gamma.ppf(0.01),
                gamma.ppf(0.99), 100)

# Función de Probabilidad
gamma_fp = gamma.pdf(x)
plt.plot(x, gamma_fp)
plt.title('Distribución Gamma')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# Generación de valores aleatorios
aleatorios = gamma.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Gamma')
plt.show()
```

**Ejemplo de aplicación**

La distribución gamma aparece cuando se realiza el estudio de la duración de elementos físicos (tiempo de vida). 

Esta distribución presenta como propiedad interesante la “falta de memoria”. Por esta razón, es muy utilizada en las teorías de la fiabilidad, mantenimiento y fenómenos de espera (por ejemplo en una consulta médica “tiempo que transcurre hasta la llegada del segundo paciente”).

Por ejemplo, el número de pacientes que llegan a la consulta de un médico sigue una distribución de Poisson de media 3 pacientes por hora. Calcular la probabilidad de que transcurra menos de una hora hasta la llegada del segundo paciente. Debe tenerse en cuenta que la variable aleatoria “tiempo que transcurre hasta la llegada del segundo paciente” sigue una distribución Gamma (6, 2).

### 6. Log-normal

```python
# Graficando Log-Normal
sigma = 0.6
lognormal = stats.lognorm(sigma)
x = np.linspace(lognormal.ppf(0.01),
                lognormal.ppf(0.99), 100)

# Función de Probabilidad
lognorm_fp = lognormal.pdf(x)
plt.plot(x, lognorm_fp)
plt.title('Distribución Log-normal')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

Histograma:

```python
# Generación de valores aleatorios
aleatorios = lognormal.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Log-normal')
plt.show()
```

**Ejemplo de aplicación**

La distribución lognormal es útil para modelar datos de numerosos estudios médicos tales como el período de incubación de una enfermedad, los títulos de anticuerpo a un virus, el tiempo de supervivencia en pacientes con cáncer o SIDA, el tiempo hasta la seroconversión de VIH+, etc.

Por ejemplo, supóngase que la supervivencia, en años, luego de una intervención quirúrgica (tiempo que pasa hasta que ocurre la muerte del enfermo) en una cierta población sigue una distribución lognormal de parámetro de escala 2,32 y de forma 0,20.
