### Carga de los datasets

```python
fifa_df = pd.read_csv('fifa_datos.csv')
empleo_df = pd.read_csv('dataset-sri-resumen-v2.csv')
```

### Estadística descriptiva

```python
# Definición de funciones
def corr_by_id(df, id_columns=[]):
    for column in id_columns:
        df_corr = df.corr()[column]
        df_corr = df_corr.sort_values(ascending=False)
        print('\nAtributo:', column.upper())
        print('\n10 Atributos con  mayor correlación\n')
        print(df_corr.iloc[:10])

        print('\n10 Atributos con menor correlación\n')
        print(df_corr.iloc[-10:])
        sliced_df = df_corr.head(10)
        df_corr = pd.concat([sliced_df, df_corr.tail(10)]).to_frame()
        plt.clf()
        print('Heatmap')
        sns.heatmap(df_corr, 
            xticklabels=df_corr.columns,
            yticklabels=df_corr.index)
        plt.show()
```

### Datos de Fifa

```python
# Descripción General por atributos
columns = ['Overall', 'Age', 'Potential', 'Skill Moves', 
       'International Reputation']
print(f'Descripción "{columns}"')
fifa_df[columns].describe()
```

**Correlación**

```python
fifa_df.corr()
```

```python
# Correlación Específica
corr_by_id(fifa_df, id_columns=['Overall', 'Age'])
```
