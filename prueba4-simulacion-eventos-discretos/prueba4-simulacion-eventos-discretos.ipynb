```python
import random
import simpy

from scipy import stats

class Vehicle():
    def __init__(self, name, big_tonnage = False):
        self.name = name
        self.big_tonnage = big_tonnage
        self.fix = False
        self.day_fix = 0
        
class Emov():
    
    def __init__(self, env):
        
        self.env = env
        self.big_tonnage_lane = simpy.Resource(self.env, 1)
        self.normal_tonnage_lane = simpy.Resource(self.env, 2)
        
        self.norm = stats.norm(MU, SIGMA)
        
        self.bernoulli_fix = stats.bernoulli(FIX_PROB)
        self.bernoulli_big_tonnage = stats.bernoulli(BIG_TONNAGE_PROB)
        
        self.vehicles_to_fix = []
        
    def get_parked_time(self):
        return random.randint(MIN_PACKED, MAX_PACKED)
    
    def get_ticket_time(self):
        return random.randint(MIN_TICKET, MAX_TICKET)
    
    def get_pay_license_plate_time(self):
        return random.randint(MIN_PAY_LICENSE_PLATE, 
                              MAX_PAY_LICENSE_PLATE)
    
    def get_revision_time(self):
        return abs(self.norm.rvs(1))
    
    def make_fix(self):
        return abs(self.bernoulli_fix.rvs(1))
    
    def is_big_tonnage(self):
        return abs(self.bernoulli_big_tonnage.rvs(1))
    
    def get_paste_stamp_time(self):
        return random.randint(MIN_PASTE_STAMP, 
                              MAX_PASTE_STAMP)
    def get_out_time(self):
        return random.randint(MIN_OUT, 
                              MAX_OUT)
        
    def sim_process(self):
        yield self.env.timeout(self.get_parked_time())
        yield self.env.timeout(self.get_ticket_time())
        yield self.env.timeout(self.get_pay_license_plate_time())
        
    def sim_revision(self, vehicle):
        yield self.env.timeout(self.get_revision_time())
        
        if self.make_fix() == 0:
            vehicle.fix = True
            vehicle.day_fix = 2
            vehicles_to_fix.append(vehicle)
        else:
            yield self.env.timeout(self.get_paste_stamp_time())
            
        yield self.env.timeout(self.get_out_time())
        print(f'Sale vehículo {vechicle.name} de gran tonelada =', 
              vechicle.big_tonnage, 'Con arreglo = ', vechicle.fix,
              self.env.now)
    
def get_fixed_vehicles(emov):
    fixed_vehicles = []
    for vechile in emov.vehicles_to_fix:
        if vechile.fix and vechile.day_fix == 0:
            fixed_vehicles.append(vechile)
        elif vechile.fix and vechile.day_fix > 0:
            vechile.day_fix -= 1

    return fixed_vehicles        
    
        
def arrival_vehicle(emov):
    
    global ticket
    
    fixed_vehicles = get_fixed_vehicles(emov)
    vehicle = None

    if len(fixed_vehicles) > 0:
        i = random.randint(0, len(fixed_vehicles))
        vehicle = fixed_vehicles.pop(i)

    if vehicle == None:
        name = 'Vehicle ' + str(ticket)
        vechicle = Vehicle(name)

        if emov.is_big_tonnage() == 1:
            vechicle.big_tonnage = True
        ticket += 1

    print(f'Se genera vehículo {vechicle.name} de gran tonelada =', vechicle.big_tonnage, env.now)

    if vehicle and vehicle.big_tonnage:
        with emov.big_tonnage_lane as big_tonnage:
            yield big_tonnage
            yield env.process(emov.sim_process())
            yield env.process(emov.sim_revision(vehicle))

    else:
        with emov.normal_tonnage_lane as normal_tonnage:
            yield normal_tonnage
            yield env.process(emov.sim_process())
            yield env.process(emov.sim_revision(vehicle))

            
def start_sim(emov):
    while True:
        yield env.timeout(WAIT_TIME_TO_ARRIVAL)
        env.process(arrival_vehicle(emov))
```

```python
MIN_PACKED = 3
MAX_PACKED = 5

MIN_TICKET = 1
MAX_TICKET = 3

MIN_PAY_LICENSE_PLATE = 5
MAX_PAY_LICENSE_PLATE = 10

MIN_PASTE_STAMP = 1
MAX_PASTE_STAMP = 3

MIN_OUT = 5
MAX_OUT = 10

MU = 20
SIGMA = 10

FIX_PROB = 0.6
BIG_TONNAGE_PROB = 0.15

WAIT_TIME_TO_ARRIVAL = 3

SIM_DAYS = 30

SIM_TIME = 3 * 60 * 2


vehicles_to_fix = []

ticket = 0



env = simpy.Environment()
emov = Emov(env)
env.process(start_sim(emov))
env.run(until=SIM_TIME)
```
